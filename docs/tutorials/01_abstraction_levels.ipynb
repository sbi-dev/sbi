# Abstraction Levels in sbi
This tutorial explains the three levels of interaction with the `sbi` toolbox.

### Level 1: High-Level (Beginner-friendly)
Automatically chooses the best default neural network.

```
from sbi.inference import NPE
inference = NPE(prior=prior)
inference = inference.append_simulations(theta, x)
density_estimator = inference.train()
posterior = inference.build_posterior()
```

### Level 2: Mid-Level (Customizing the Network)
Allows choosing specific architectures like MDN or MAF.

```
from sbi.utils import posterior_nn
custom_net = posterior_nn(model='mdn', hidden_features=50)
inference = NPE(prior=prior, density_estimator=custom_net)
density_estimator = inference.train(learning_rate=0.001)
```

### Level 3: Low-Level (Full Control)
Manual control over training loops and density estimators.

```
density_estimator = inference.train(force_first_round_loss=True)
posterior = inference.build_posterior(density_estimator)
```
